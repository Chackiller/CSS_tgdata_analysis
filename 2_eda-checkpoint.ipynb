# Telegram users activity chart by day of the week

### Questions

- How the activity of people on the network changes depending on the day of the week
- How my activity on the network changes depending on the day of the week
- Comparing my online activity with the average user activity depending on the day of the week

### Conclusion

- My online activity during the week strongly coincides with the statistics of the average telegram user
- The activity of telegram users falls on weekends. This may be due to the fact that on weekends people are more likely to rest, see friends and work less
- The slight peak of my activity on Thursday may be due to the discussion of plans for Friday night or weekend and the desire to complete all work before the end of the work week


```python
newDate = df.date.dt.day_name()
weekdf = pd.DataFrame( columns = ['Days', 'Number'])
weekdf = weekdf.append({'Days' : 'Monday', 'Number' : np.where(newDate == 'Monday', 1, 0).sum()},  ignore_index=True)
weekdf = weekdf.append({'Days' : 'Tuesday', 'Number' : np.where(newDate == "Tuesday", 1, 0).sum()},  ignore_index=True)
weekdf = weekdf.append({'Days' : 'Wednesday', 'Number' : np.where(newDate == "Wednesday", 1, 0).sum()},  ignore_index=True)
weekdf = weekdf.append({'Days' : 'Thursday', 'Number' : np.where(newDate == "Thursday", 1, 0).sum()},  ignore_index=True)
weekdf = weekdf.append({'Days' : 'Friday', 'Number' : np.where(newDate == "Friday", 1, 0).sum()},  ignore_index=True)
weekdf = weekdf.append({'Days' : 'Saturday', 'Number' : np.where(newDate == "Saturday", 1, 0).sum()},  ignore_index=True)
weekdf = weekdf.append({'Days' : 'Sunday', 'Number' : np.where(newDate == "Sunday", 1, 0).sum()},  ignore_index=True)

fig, ax = plt.subplots()
make_beautifull(fig, ax)

plt.xlabel('Days of week', fontsize = 18, labelpad=15 )
plt.ylabel('Number of messages and posts', fontsize = 18, labelpad=15)
plt.title('The number of messages sent by Telegram users', fontsize = 32, weight='bold', pad=15)

plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'

plt.bar(weekdf.Days, weekdf.Number, color = "#2CBDFE", width = 0.5)
```

```python
import wordcloud 
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
from PIL import Image

def readFile(fileName):
    with open(fileName, 'r', errors="ignore", encoding='cp1251') as line:
        words = line.read().splitlines()
    return words

words = []
stopwords_ru = stopwords.words('russian')
stopwords_ua = readFile('C:/Users/minep/Desktop/test3.txt')

def get_words(data):
    for ind in data['message'].index:
        for word in str(data['message'][ind]).split():
            word.replace(".","")
            if (word.lower() not in stopwords_ua) & (word.lower() not in stopwords_ru) & (word.lower() != 'nan') & (word.isalpha()) & (len(word)>2):
                words.append(word.lower())
    return words
```
